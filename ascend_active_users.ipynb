```python
# Import the necessary modules

import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical computations

import matplotlib.pyplot as plt  # For plotting
import seaborn as sns  # For statistical data visualization
```

```python
# Load the dataset

ascend_active_users = pd.read_csv('C:/Users/BrindhaGanesan/OneDrive - IMMIGRANT EMPLOYMENT COUNCIL OF BC SOCIETY/Documents/ASCEND/Datasets/ascend_active_users.csv')
ascend_active_users.head()
```

```python
# 'ascend_active_users' is our DataFrame
department_counts = ascend_active_users['DEPARTMENT'].value_counts()

# Creating the bar chart using matplotlib
plt.figure(figsize=(10, 6))
department_counts.plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Number of Active Users by Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the chart
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get the top 5 departments
top_5_departments = ascend_active_users['DEPARTMENT'].value_counts().head(5)

# Creating the bar chart
plt.figure(figsize=(8, 5))
top_5_departments.plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Top 5 Departments by Active Users', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the chart
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
